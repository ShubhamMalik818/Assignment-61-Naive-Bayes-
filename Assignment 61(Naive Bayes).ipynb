```python
Q1. What is Bayes theorem?

ANS- Bayes theorem is a formula in probability theory that describes the probability of an event occurring given that another event has 
     already occurred. It is named after Thomas Bayes, an English mathematician who first published it in 1763.

The theorem is often used in machine learning and data science to update the probability of a hypothesis given new evidence. 
For example, if you are trying to predict whether a patient has a disease, you can use Bayes theorem to update the probability of the 
patient having the disease given the results of a medical test.

The formula for Bayes theorem is:

P(A|B) = P(B|A)P(A) / P(B)

where:

P(A|B) is the probability of event A occurring given that event B has already occurred
P(B|A) is the probability of event B occurring given that event A has already occurred
P(A) is the probability of event A occurring
P(B) is the probability of event B occurring


Bayes theorem can be used to solve a variety of problems in machine learning and data science. For example, it can be used to:

1. Classify data into different categories
2. Predict the outcome of an event
3. Make decisions based on incomplete or uncertain information

Bayes theorem is a powerful tool that can be used to improve the accuracy of machine learning models. However, it is important to use 
it correctly, as it can be easy to make mistakes.

Here are some examples of how Bayes theorem can be used in machine learning and data science:

1. Classification: Bayes theorem can be used to classify data into different categories. For example, you could use Bayes theorem to classify 
   images into different categories, such as cat, dog, or flower.

2. Prediction: Bayes theorem can be used to predict the outcome of an event. For example, you could use Bayes theorem to predict whether 
               a patient has a disease based on the results of a medical test.
3. Decision making: Bayes theorem can be used to make decisions based on incomplete or uncertain information. For example, you could use 
                    Bayes theorem to decide whether to take a risk based on the probability of the risk occurring.
```

```python
Q3. What is the relationship between Bayes theorem and conditional probability?

ANS- Bayes theorem and conditional probability are closely related concepts in probability theory. Bayes theorem can be used to calculate 
     the posterior probability of an event given new evidence, while conditional probability is the probability of one event occurring 
     given that another event has already occurred.

The relationship between Bayes theorem and conditional probability can be seen in the formula for Bayes theorem:

       P(A|B) = P(B|A)P(A) / P(B)

where:

P(A|B) is the posterior probability of event A occurring given that event B has already occurred
P(B|A) is the likelihood of event B occurring given that event A has already occurred
P(A) is the prior probability of event A occurring
P(B) is the marginal probability of event B occurring

The posterior probability, P(A|B), is the probability of event A occurring after we have observed event B. The likelihood, P(B|A), is 
the probability of event B occurring given that event A has already occurred. The prior probability, P(A), is the probability of event 
A occurring before we have observed event B. The marginal probability, P(B), is the probability of event B occurring regardless of whether 
event A has occurred.

As you can see, the posterior probability can be calculated by multiplying the likelihood by the prior probability and then dividing by the 
marginal probability. This is the same as the definition of conditional probability.

Therefore, Bayes theorem can be seen as a way of calculating conditional probability. In fact, Bayes theorem is often used to update the 
prior probability of an event given new evidence. This is because the likelihood of an event occurring can often be determined from data, 
while the prior probability is often subjective.

Bayes theorem is a powerful tool that can be used to improve the accuracy of predictions by taking into account prior knowledge. It is a 
versatile technique that can be applied to a wide range of problems.
```

```python
Q5. You have a dataset with two features, X1 and X2, and two possible classes, A and B. You want to use Naive Bayes to classify a new 
    instance with features X1 = 3 and X2 = 4. The following table shows the frequency of each feature value for each class:
    
    Class  X1=1  X1=2  X1=3  X2=1  X2=2  X2=3  X2=4
    A       3     3     4     4     3     3      3
    B       2     2     1     2     2     2      3

Assuming equal prior probabilities for each class, which class would Naive Bayes predict the new instance to belong to?
```

```python
import numpy as np

# Calculate the prior probabilities
prior_probability_a = 0.5
prior_probability_b = 0.5

# Calculate the likelihoods
likelihood_a_x1_3 = 4 / 13
likelihood_a_x2_4 = 3 / 13
likelihood_b_x1_3 = 1 / 13
likelihood_b_x2_4 = 3 / 13

# Calculate the posterior probabilities
posterior_probability_a = prior_probability_a * likelihood_a_x1_3 * likelihood_a_x2_4
posterior_probability_b = prior_probability_b * likelihood_b_x1_3 * likelihood_b_x2_4

# Print the posterior probabilities
print(posterior_probability_a)
print(posterior_probability_b)
```
